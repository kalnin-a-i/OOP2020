## Двусвязный список 

```python
class Node():
    prev = None
    value = None
    next = None
    
    def __init__(self, new_prev=None, new_value=None, new_next=None):
        self.prev = new_prev
        self.value = new_value
        self.next = new_next

        
class DoubleLinkedList():
    first = None
    last = None
    len = 0
    
    def __init__(self, new_list=None):
        if new_list == None:
            return
        self.len = len(new_list)
        self.first = Node(None, new_list[0], None)
        cur = self.first
        for i in range(1, len(new_list)):
            cur.next = Node(cur, new_list[i], None)
            cur = cur.next
        self.last = cur
        return
    
    def __str__(self):
        cur = self.first
        out = ''
        while cur != None:
            if cur == self.first:
                out += str(cur.value)
                cur = cur.next
            else:
                out += ', ' + str(cur.value)
                cur = cur.next
        return out
    
    #Добавление в начало
    def push(self, x):
        if self.first != None:
            self.first = Node(None, x, self.first)
            self.first.next.prev = self.first
        else:
            self.first = Node(None, x, None)
            self.last = self.first
        self.len += 1
        return
    
    # Добавление в i позицию
    def insert(self, x, i):
        if self.first == None:
            self.first = Node(None, x, None)
            self.last = self.first
            self.len += 1
            return
        if i == 1:
            self.first = Node(None, x, self.first)
            self.first.next.prev = self.first
            self.len += 1
            return
        if i == self.len:
            self.last.next = Node(self.last, x, None)
            self.last = self.last.next
            self.len += 1
            return
        cur = self.first
        for j in range(i - 2):
            cur = cur.next
        cur.next = Node(cur, x, cur.next)
        cur.next.next.prev = cur.next
        self.len += 1 
        return
    
    #Удаление из начала
    def pop(self):
        assert self.first is not None, 'List is empty'
        x = self.first.value
        self.first = self.first.next
        self.first.prev = None
        self.len -= 1
        return x
    
    #Удаление из произвольного места
    def delete(self, i):
        assert self.first is not None, 'list empty'
        if i > self.len:
            raise IndexError('List index out of range')
        if i == 0:
            x = self.first.value
            self.first = self.first.next
            self.first.prev = None
            self.len -= 1
            return x
        if i == self.len:
            x = self.last.value
            self.last = self.last.prev
            self.last.next = None
            self.len -= 1
            return x
        cur = self.first 
        for j in range(i - 2):
            cur = cur.next
        x = cur.next.value
        cur.next = cur.next.next
        cur.next.prev = cur
        self.len -= 1
        return x
    
    #Поиск по значению
    def search(self, value):
        assert self.first is not None, 'list empty'
        cur = self.first
        counter = 0
        while cur != None:
            if cur.value == value:
                return counter
            else:
                counter += 1
                cur = cur.next
        if cur == None:
            return "No such element"
    # Итератор
    def __iter__(self):
        self.__curr = self.first
        return self
    
    def __next__(self):
        if self.__curr == None:
            raise StopIteration()
        x = self.__curr.value
        self.__curr = self.__curr.next
        return x
    
    def __len__(self):
        return self.len
    
    def __getitem__(self, i):
        if i >= self.len:
            raise IndexError("Index out of range")
        cur = self.first
        for j in range(i):
            cur = cur.next
        return cur.value
```

### Декоратор оборачиват вывод в квадратный скобки

```python
def decorator(function):
    def wrapper(a):
        print('[', end='')
        function(a)
        print(']')  
    return wrapper

@decorator
def my_print(list):
    print(list, end=' ')
```
